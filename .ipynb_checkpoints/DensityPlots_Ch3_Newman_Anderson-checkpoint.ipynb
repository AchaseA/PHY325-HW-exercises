```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#exercise 3.1 from Newman

data = np.loadtxt("sunspots.txt", float)
```

```python
x_data = data[:,0]  #time
y_data = data[:,1]  #number of sunspots for the given month
r = 5
i = 6
ave_month = []
loc_average = []

for q in range(5,995):                             #using a for loop to make a running average for the data points
    loc_average.append(np.sum(y_data[i-6:i+5] /11))
    ave_month.append(i)
    i = i + 1
    
plt.plot(ave_month,loc_average, "k.")
plt.plot(x_data, y_data, "r")
plt.xlim(0,1000)
plt.xlabel("Months")
plt.ylabel("# of sunspots")
```
